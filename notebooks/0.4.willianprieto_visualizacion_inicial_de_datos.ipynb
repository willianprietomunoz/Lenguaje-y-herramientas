```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#scatter plot (diagrama de dispersion)
from cProfile import label
from turtle import title


x = np.arange(1,20)
y = x**2
y1 = x**2 + 5*x

plt.scatter(x,y, color='green', marker='s', label='x**2', s=30)
plt.scatter(x,y1, color='red', marker='o', label='x**2+5x', s=30)
plt.legend(title='Funciones')
plt.title('Tipos de funciones', fontsize=20)
plt.xlabel('X')
plt.ylabel('f(x)')
#plt.xlim(0,5)
plt.grid()
plt.text(5,300, 'este texto', color='blue')

# agregar coordenadas de maracadors en texto

for val_x in x:
    plt.text(val_x,val_x**2, f'({val_x},{val_x**2})')
    
```

```python
# Histograma
media = 20
std = 3
numeros = np.random.normal(media, std, 1000)
plt.hist(numeros,bins=20, cumulative=False, density=True);
plt.title('Distribucion normal')
plt.xlabel('Numeros')
plt.ylabel('Frecuencia')
```
